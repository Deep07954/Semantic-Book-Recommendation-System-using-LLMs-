```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()
sns.heatmap(df.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np 
df["missing_description"]=np.where(df["description"].isna(),1,0)
df["age_of_books"]=2024-df["published_year"]
```

```python
columns_of_interest=["average_rating","num_pages","age_of_books","missing_description"]
correlation_matrix=df[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
df_missing=df[(~df["missing_description"].isna()) &
(~df["average_rating"].isna()) &
(~df["num_pages"].isna())      &
(~df["published_year"].isna())]
df_missing.shape
```

```python
# Filter out invalid or negative word counts if any exist
df_missing = df_missing[df_missing['words_in_description'] >= 0]

# Plot histogram
plt.figure(figsize=(10,6))
plt.hist(df_missing['words_in_description'], bins=30, color='brown', edgecolor='black')

plt.title('Distribution of Words in Description', fontsize=14)
plt.xlabel('Number of Words in Description', fontsize=12)
plt.ylabel('Count (Number of Books)', fontsize=12)
plt.grid(axis='y', alpha=0.4)

plt.show()
```

```python
df_missing_25_words = df_missing[df_missing["words_in_description"] >= 25]
```

```python
df_missing_25_words
```

```python
mask = df_missing_25_words["words_in_description"] >= 25
df_missing_25_words.loc[mask, "title_and_subtitle"] = (np.where(df_missing_25_words["subtitle"].isna(), df_missing_25_words["title"],
            df_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
mask = df_missing_25_words["words_in_description"] >= 25

df_missing_25_words.loc[mask, "title_and_subtitle"] = np.where(
    df_missing_25_words.loc[mask, "subtitle"].isna(),
    df_missing_25_words.loc[mask, "title"],
    df_missing_25_words.loc[mask, ["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)
```

```python
df_missing_25_words
```

```python
mask = df_missing_25_words["words_in_description"] >= 25
df_missing_25_words.loc[mask, "tagged_description"] = (
    df_missing_25_words.loc[mask, ["isbn13", "description"]]
    .astype(str)
    .agg(" ".join, axis=1)
)
```

```python
df_missing_25_words
```

```python
df_missing_25_words.drop(["subtitle","words_in_description","missing_description","age_of_books"],axis=1).to_csv("books_cleaned.csv", index=False)
```
